```python
def rodriguez_formula(axis, theta):
    """
    Construct a 3D rotation matrix for rotation by angle theta around a given axis.
    
    Parameters:
    axis : array-like, shape (3,)
        The axis of rotation as a 3D vector [x, y, z]
    theta : float
        The angle of rotation in radians
        
    Returns:
    R : numpy.ndarray, shape (3, 3)
        The 3x3 rotation matrix
    """
    # Convert axis to numpy array and normalize it
    axis = np.array(axis, dtype=float)
    axis = axis / np.linalg.norm(axis)
    
    # Extract components
    ux, uy, uz = axis
    
    # Compute trigonometric values
    cos_theta = np.cos(theta)
    sin_theta = np.sin(theta)
    one_minus_cos = 1 - cos_theta
    
    # Construct rotation matrix using Rodrigues' formula
    R = np.array([
        [cos_theta + ux*ux*one_minus_cos,     ux*uy*one_minus_cos - uz*sin_theta,  ux*uz*one_minus_cos + uy*sin_theta],
        [uy*ux*one_minus_cos + uz*sin_theta,  cos_theta + uy*uy*one_minus_cos,     uy*uz*one_minus_cos - ux*sin_theta],
        [uz*ux*one_minus_cos - uy*sin_theta,  uz*uy*one_minus_cos + ux*sin_theta,  cos_theta + uz*uz*one_minus_cos]
    ])
    
    return R
```

```python
rotation_axis = np.cross(bc_vsc, [0, 0, 1])
rot_angle = np.arccos(bc_vsc[-1]) # the dot product


rod_mat = rodriguez_formula(rotation_axis, rot_angle)
misalignet_mountconfig = rod_mat.T@perfect_mountconfig
```
